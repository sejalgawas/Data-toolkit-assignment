```python
# How do you create a 2D NumPy array and calculate the sum of each row

import numpy as np

A = np.array([[10, 20, 30],
              [40, 50, 60],
              [70, 80, 90]])

print(A)

row_sums = np.sum(A, axis=1)

print("\nSum of each row:")
print(row_sums)
```

```python
#How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 75000, 80000, 85000],
    'Experience': [1, 3, 5, 7, 9]
}
df = pd.DataFrame(data)

corr_matrix = df.corr()

print("Correlation Matrix:")
print(corr_matrix)

plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Generate a bar plot using Plotly

import plotly.express as px
import pandas as pd
data = {
    'Fruit': ['Apple', 'Banana', 'Orange', 'Mango'],
    'Quantity': [10, 15, 7, 12]
}
df = pd.DataFrame(data)

fig = px.bar(df, x='Fruit', y='Quantity', title="Fruit Quantity Bar Plot")

fig.show()
```

```python
#Create a DataFrame and add a new column based on an existing column
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

df['Age_in_5_years'] = df['Age'] + 5
df['Category'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Experienced')

print("\nDataFrame after adding new columns:")
print(df)

```

```python
# Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5,]
y3 = [2, 3, 4, 5, 6]

plt.plot(x, y1, label="Line 1", marker='o')
plt.plot(x, y2, label="Line 2", marker='s')
plt.plot(x, y3, label="Line 3", marker='^')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Multiple Lines")

plt.legend()

plt.show()
```

```python
#Generate a Pandas DataFrame and filter rows where a column value is greater than a thresholdA
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 32, 29, 41, 19],
    'Salary': [50000, 60000, 55000, 80000, 40000]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

threshold = 30

filtered_df = df[df['Age'] > threshold]

print("\nFiltered DataFrame (Age > 30):")
print(filtered_df)
```

```python
#Create a histogram using Seaborn to visualize a distributionA
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=50, scale=10, size=200)

sns.histplot(data, bins=20, kde=True, color="skyblue")

plt.title("Histogram of Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
#Perform matrix multiplication using NumPy

import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

result = np.dot(A, B)   # OR result = A @ B

print("\nResult of Matrix Multiplication:\n", result)
```

```python
#A Create a 3D scatter plot using Plotly

import plotly.express as px
import pandas as pd

data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 11, 12, 13, 14],
    'Z': [5, 6, 7, 8, 9],
    'Category': ['A', 'A', 'B', 'B', 'C']
}
df = pd.DataFrame(data)

fig = px.scatter_3d(df, x='X', y='Y', z='Z',
                    color='Category',
                    size_max=10,
                    title="3D Scatter Plot Example")
fig.show()
```
